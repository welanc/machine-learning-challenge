```python
import pandas as pd
```

```python
data = pd.read_csv("../data/cumulative.csv")

df = pd.DataFrame(data)

df.head()
```

```python
# initializing list
cols = df.columns.tolist()
print(cols)
```

```python
# Python code to find strings with substrings
# using filter() + lambda

# initializing substring
subs = 'err'

# using filter() + lambda
# to get string with substring
res = list(filter(lambda x: subs in x, cols))

# printing result
print (str(res))
```

```python
no_err_df = df.drop(columns=res)
no_err_df
```
