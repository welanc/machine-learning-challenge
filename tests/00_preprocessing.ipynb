```python
import pandas as pd
```

```python
data = pd.read_csv("../data/cumulative.csv")

df = pd.DataFrame(data)

df.head()
```

```python
df.count()
```

```python
# Python code to find strings with substrings
# using filter() + lambda

# initializing list
cols = df.columns.tolist()
print(cols)
```

```python
# initializing substring
subs = 'err'

# using filter() + lambda
# to get string with substring
res = list(filter(lambda x: subs in x, cols))

# printing result
print (str(res))
```

```python
no_err_df = df.drop(columns=res)
no_err_df
```

```python
no_err_df.columns
```

```python
keep_cols = [
    "kepid",
    "kepoi_name",
    "koi_disposition",
    "koi_pdisposition",
    "koi_score",
    "koi_fpflag_nt",
    "koi_fpflag_ss",
    "koi_fpflag_co",
    "koi_fpflag_ec",
    "koi_period",
    "koi_impact",
    "koi_duration",
    "koi_depth",
    "koi_prad",
    "koi_teq",
    "koi_model_snr",
    "koi_steff",
    "koi_slogg",
    "koi_srad"
]
```

```python
reduced_df = no_err_df[keep_cols]
reduced_df
```
